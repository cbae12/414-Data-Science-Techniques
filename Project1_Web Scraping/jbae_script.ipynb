```python
# Import pandas package
import pandas as pd

# Retrieve given data set from "dataset_1.csv", then assign each column to a list
df=pd.read_csv("dataset_1.csv")
urls=df["lyrics url"]
yrs=df["year"]
names=df["artist/band name"]
titles=df["song title"]
```

```python
# Import packages
from bs4 import BeautifulSoup
import requests
import time
import random
```

```python
# Create variables
all_lyrics=[]
bad_lyrics=[]
count=0

# Iterate through the list of URLs using for loop to get lyrics for each song
for url in urls:
    cnt=1
    song_lyrics=''
    # Assign GET request from URL to a variable
    page=requests.get(url)
    # Check status of the request. 200 is OK.
    stat=page.status_code
    print(f"Status Code: {stat}\n")
    if stat != 200:
        all_lyrics.append(None)
        bad_lyrics.append(count+1)
        count+=1
        continue
    # Parses content of the request made in lxml format
    soup=BeautifulSoup(page.content, "lxml")
    # Converts Beautiful Soup pase tree into a nicely formatted Unicode string
    soup.prettify()
    # Finds all matches with search on class "col-xs-12 col-lg-8 text-center"
    x=soup.find_all(class_="col-xs-12 col-lg-8 text-center")
    # Finds the first match with searches on "div" and has no class
    lyrics=x[0].find("div", class_=None)
    # Converts into list of each lyric's line in text format in precedence
    just_lyrics=lyrics.text.strip().split("\n")
    # Iterates over the list of lines from lyric to concatenate them into a string variable
    for line in just_lyrics:
        song_lyrics+=f"{line}\n"
        if cnt > len(just_lyrics):
            song_lyrics+=f"{line}"
        cnt+=1
    print(f"Song#{count+1}: {titles[count]}\n\n{song_lyrics}\n\n")
    # Appends string variable for lyrics into a list of all lyrics in the given data set
    all_lyrics.append(song_lyrics)
    count+=1
    # Delays each loop execution by non-integer value that is between 2 and 4 seconds
    time.sleep(random.uniform(30,60))
```

```python
print(f"These are the URLs with error:\n\n{bad_lyrics}\n\n")
```

```python
# Creates a new column with a list of lyrics to be the data set for that column
df['song_lyrics']=all_lyrics
```

```python
df.to_csv('dataset_1_revised.csv')
```
